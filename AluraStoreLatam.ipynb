### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda.info()
```

```python
# Caluclo para el totoal de ingreso de la tienda 1 
total_Ingreso_x_Precio_tienda1 = tienda["Precio"].sum()
total_CostE_tienda1 = tienda["Costo de envío"].sum()
Utilidad_tienda1 = total_Ingreso_x_Precio_tienda1 - total_CostE_tienda1
print(f'El total de ingresos por venta que tuvo la tieneda 1 fue: ${total_Ingreso_x_Precio_tienda1}') 

# Caluclo para el totoal de ingreso de la tienda 2
total_Ingreso_x_Precio_tienda2 = tienda2["Precio"].sum()
total_CostE_tienda2 = tienda2["Costo de envío"].sum()
Utilidad_tienda2 = total_Ingreso_x_Precio_tienda2 - total_CostE_tienda2
print(f'El total de ingresos por venta que tuvo la tieneda 2 fue: ${total_Ingreso_x_Precio_tienda2}') 

# Caluclo para el totoal de ingreso de la tienda 3 
total_Ingreso_x_Precio_tienda3 = tienda3["Precio"].sum()
total_CostE_tienda3 = tienda3["Costo de envío"].sum()
Utilidad_tienda3 = total_Ingreso_x_Precio_tienda3 - total_CostE_tienda3
print(f'El total de ingresos por venta que tuvo la tieneda 3 fue: ${total_Ingreso_x_Precio_tienda3}') 

# Caluclo para el totoal de ingreso de la tienda 4
total_Ingreso_x_Precio_tienda4 = tienda4["Precio"].sum()
total_CostE_tienda4 = tienda4["Costo de envío"].sum()
Ingreso_tienda4 = total_Ingreso_x_Precio_tienda4 - total_CostE_tienda4
print(f'El total de ingresos por venta que tuvo la tieneda 4 fue: ${total_Ingreso_x_Precio_tienda4}') 
```

```python
df_Fac = pd.DataFrame()
df_Fac['Precio T1'] = tienda.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T1'] = tienda.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T1'] = df_Fac['Precio T1'] - df_Fac['Costo Envio T1']
df_Fac['Precio T2'] = tienda2.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T2'] = tienda2.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T2'] = df_Fac['Precio T2'] - df_Fac['Costo Envio T2']
df_Fac['Precio T3'] = tienda3.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T3'] = tienda3.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T3'] = df_Fac['Precio T3'] - df_Fac['Costo Envio T3']
df_Fac['Precio T4'] = tienda4.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T4'] = tienda4.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T4'] = df_Fac['Precio T4'] - df_Fac['Costo Envio T4']
df_Fac
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_totales = pd.DataFrame({
    'Ingresos_x_Precio Tienda 1': [df_Fac['Precio T1'].sum()],
    'Ingresos_x_Precio Tienda 2': [df_Fac['Precio T2'].sum()],
    'Ingresos_x_Precio Tienda 3': [df_Fac['Precio T3'].sum()],
    'Ingresos_x_Precio Tienda 4': [df_Fac['Precio T4'].sum()]
})
df_totales
```

```python
import matplotlib.pyplot as plt

# Transponer df_totales para que los nombres de las tiendas sean las categorías del eje X
income_series = df_totales.iloc[0]
income_series.index = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 8))

# Generar el gráfico de columnas (barras)
income_series.plot(kind='bar', ax=ax, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'], width=0.4, label='Ingresos por Columna')

# Generar el gráfico de línea sobre los mismos ejes
income_series.plot(kind='line', ax=ax, color='black', marker='o', linestyle='--', label='Ingresos por Línea')

ax.set_title('Ingresos Totales por Tienda (Columnas y Línea)')
ax.set_xlabel('Tienda')
ax.set_ylabel('Ingresos')
ax.legend() # Mostrar la leyenda para diferenciar los elementos
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
df_VPC = pd.DataFrame()
df_VPC['Total Ventas T1'] = tienda.groupby('Categoría del Producto')['Precio'].count()
df_VPC['Total Ventas T2'] = tienda2.groupby('Categoría del Producto')['Precio'].count()
df_VPC['Total Ventas T3'] = tienda3.groupby('Categoría del Producto')['Precio'].count()
df_VPC['Total Ventas T4'] = tienda4.groupby('Categoría del Producto')['Precio'].count()
df_VPC['Total Categoria'] = df_VPC['Total Ventas T1'] + df_VPC['Total Ventas T2'] + df_VPC['Total Ventas T3'] + df_VPC['Total Ventas T4']
df_VPC
```

```python
df_VPC.sort_values(by='Total Categoria', ascending=False)
```

```python
df_VPC['Total Categoria'].plot(kind='barh',figsize=(12,8),color='purple', title='Ventas por Categoria')
```

# 3. Calificación promedio de la tienda


```python
df_Client = pd.DataFrame()
df_Client['Calificación Promedio T1'] = tienda.groupby('Producto')['Calificación'].mean()
df_Client['Calificación Promedio T2'] = tienda2.groupby('Producto')['Calificación'].mean()
df_Client['Calificación Promedio T3'] = tienda3.groupby('Producto')['Calificación'].mean()
df_Client['Calificación Promedio T4'] = tienda4.groupby('Producto')['Calificación'].mean()
df_Client['Satisfacción promedio total por producto'] = (df_Client['Calificación Promedio T1'] + df_Client['Calificación Promedio T2'] + df_Client['Calificación Promedio T3'] + df_Client['Calificación Promedio T4'])/4
df_Client
```

```python
df_Client.sort_values(by='Satisfacción promedio total por producto', ascending=False)
```

```python
# Calcular los 10 productos con mayor satisfacción promedio
top_10_productos = df_Client.sort_values(by='Satisfacción promedio total por producto', ascending=False).head(10)

# Generar el gráfico de pastel para los 10 productos con mayor satisfacción promedio
ax = top_10_productos['Satisfacción promedio total por producto'].plot(
    kind='pie',
    figsize=(12, 8),
    autopct='%.2f%%',
    startangle=90,
    cmap='viridis',
    title='Top 10 Productos con Mayor Satisfacción Promedio Total'
)
ax.set_ylabel('') # Eliminar la etiqueta del eje Y que pandas añade por defecto
```

# 5. Envío promedio por tienda

```python
# tienda 1 
total_CostE_tienda1 = tienda["Costo de envío"].sum()
total_datos_Colum_CostE_tienda1 = tienda["Costo de envío"].count()
promedio_CostE_tienda1 = total_CostE_tienda1 / total_datos_Colum_CostE_tienda1
print(f'El promedio de gastos por Costo de Envio que tuvo la tieneda 1 fue: ${promedio_CostE_tienda1:.2f}') 

# tienda 2
total_CostE_tienda2 = tienda2["Costo de envío"].sum()
total_datos_Colum_CostE_tienda2 = tienda2["Costo de envío"].count()
promedio_CostE_tienda2 = total_CostE_tienda2 / total_datos_Colum_CostE_tienda2
print(f'El promedio de gastos por Costo de Envio que tuvo la tieneda 2 fue: ${promedio_CostE_tienda2:.2f}') 


# tienda 3 
total_CostE_tienda3 = tienda3["Costo de envío"].sum()
total_datos_Colum_CostE_tienda3 = tienda3["Costo de envío"].count()
promedio_CostE_tienda3 = total_CostE_tienda3 / total_datos_Colum_CostE_tienda3
print(f'El promedio de gastos por Costo de Envio que tuvo la tieneda 3 fue: ${promedio_CostE_tienda3:.2f}') 

# tienda 4
total_CostE_tienda4 = tienda4["Costo de envío"].sum()
total_datos_Colum_CostE_tienda4 = tienda4["Costo de envío"].count()
promedio_CostE_tienda4 = total_CostE_tienda4 / total_datos_Colum_CostE_tienda4
print(f'El promedio de gastos por Costo de Envio que tuvo la tieneda 1 fue: ${promedio_CostE_tienda4:.2f}') 
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_totalE = pd.DataFrame({
    'Costo de Envio Tienda 1': [(df_Fac['Costo Envio T1'].sum())/tienda["Costo de envío"].count()],
    'Costo de Envio Tienda 2': [(df_Fac['Costo Envio T2'].sum())/tienda2["Costo de envío"].count()],
    'Costo de Envio Tienda 3': [(df_Fac['Costo Envio T2'].sum())/tienda3["Costo de envío"].count()],
    'Costo de Envio Tienda 4': [(df_Fac['Costo Envio T2'].sum())/tienda4["Costo de envío"].count()]
})
df_totalE
```

```python
import matplotlib.pyplot as plt

# Transponer df_totales para que los nombres de las tiendas sean las categorías del eje X
income_series = df_totalE.iloc[0]
income_series.index = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 8))

# Generar el gráfico de columnas (barras)
income_series.plot(kind='bar', ax=ax, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'], width=0.4, label='Costo Promedio de Envio por Tienda')

# Generar el gráfico de línea sobre los mismos ejes
income_series.plot(kind='line', ax=ax, color='black', marker='o', linestyle='--', label='Costo Promedio de Envio por Tienda')

ax.set_title('Costo Promedio de Envio por Tienda (Columnas y Línea)')
ax.set_xlabel('Tienda')
ax.set_ylabel('Costo de Envio Promedio')
ax.legend() # Mostrar la leyenda para diferenciar los elementos
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
