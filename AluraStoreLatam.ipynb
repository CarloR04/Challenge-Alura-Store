````python
# Caluclo para el totoal de ingreso de la tienda 1 
total_Ingreso_x_Precio_tienda1 = tienda["Precio"].sum()
total_CostE_tienda1 = tienda["Costo de envío"].sum()
Utilidad_tienda1 = total_Ingreso_x_Precio_tienda1 - total_CostE_tienda1
print(f'El total de ingresos por venta que tuvo la tieneda 1 fue: ${total_Ingreso_x_Precio_tienda1}') 

# Caluclo para el totoal de ingreso de la tienda 2
total_Ingreso_x_Precio_tienda2 = tienda2["Precio"].sum()
total_CostE_tienda2 = tienda2["Costo de envío"].sum()
Utilidad_tienda2 = total_Ingreso_x_Precio_tienda2 - total_CostE_tienda2
print(f'El total de ingresos por venta que tuvo la tieneda 2 fue: ${total_Ingreso_x_Precio_tienda2}') 

# Caluclo para el totoal de ingreso de la tienda 3 
total_Ingreso_x_Precio_tienda3 = tienda3["Precio"].sum()
total_CostE_tienda3 = tienda3["Costo de envío"].sum()
Utilidad_tienda3 = total_Ingreso_x_Precio_tienda3 - total_CostE_tienda3
print(f'El total de ingresos por venta que tuvo la tieneda 3 fue: ${total_Ingreso_x_Precio_tienda3}') 

# Caluclo para el totoal de ingreso de la tienda 4
total_Ingreso_x_Precio_tienda4 = tienda4["Precio"].sum()
total_CostE_tienda4 = tienda4["Costo de envío"].sum()
Ingreso_tienda4 = total_Ingreso_x_Precio_tienda4 - total_CostE_tienda4
print(f'El total de ingresos por venta que tuvo la tieneda 4 fue: ${total_Ingreso_x_Precio_tienda4}') 
````

````python
df_Fac = pd.DataFrame()
df_Fac['Precio T1'] = tienda.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T1'] = tienda.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T1'] = df_Fac['Precio T1'] - df_Fac['Costo Envio T1']
df_Fac['Precio T2'] = tienda2.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T2'] = tienda2.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T2'] = df_Fac['Precio T2'] - df_Fac['Costo Envio T2']
df_Fac['Precio T3'] = tienda3.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T3'] = tienda3.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T3'] = df_Fac['Precio T3'] - df_Fac['Costo Envio T3']
df_Fac['Precio T4'] = tienda4.groupby('Categoría del Producto')['Precio'].sum()
df_Fac['Costo Envio T4'] = tienda4.groupby('Categoría del Producto')['Costo de envío'].sum()
df_Fac['Utilidad T4'] = df_Fac['Precio T4'] - df_Fac['Costo Envio T4']
df_Fac
````

````python
pd.options.display.float_format = '{:.2f}'.format
df_totales = pd.DataFrame({
    'Ingresos_x_Precio Tienda 1': [df_Fac['Precio T1'].sum()],
    'Ingresos_x_Precio Tienda 2': [df_Fac['Precio T2'].sum()],
    'Ingresos_x_Precio Tienda 3': [df_Fac['Precio T3'].sum()],
    'Ingresos_x_Precio Tienda 4': [df_Fac['Precio T4'].sum()]
})
df_totales
````

````python
import matplotlib.pyplot as plt

# Transponer df_totales para que los nombres de las tiendas sean las categorías del eje X
income_series = df_totales.iloc[0]
income_series.index = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 8))

# Generar el gráfico de columnas (barras)
income_series.plot(kind='bar', ax=ax, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'], width=0.4, label='Ingresos por Columna')

# Generar el gráfico de línea sobre los mismos ejes
income_series.plot(kind='line', ax=ax, color='black', marker='o', linestyle='--', label='Ingresos por Línea')

ax.set_title('Ingresos Totales por Tienda (Columnas y Línea)')
ax.set_xlabel('Tienda')
ax.set_ylabel('Ingresos')
ax.legend() # Mostrar la leyenda para diferenciar los elementos
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
````

Estas áreas están indicadas por colores más cálidos, lo que significa una mayor densidad de ventas.
